# Product Revenue by Year and Region

## Description

- Produce a table with the following columns:
    - ```productCode```
    - ```totalSales``` for the ```productCode```
    - Grouped by
        - ```country``` of customer.
        - ```quarter```
        - ```year```
        
        
- Let's look at the raw data ... ...

````python
%%sql

select orderNumber, orderDate, customerNumber, productCode, quantityOrdered, priceEach
    from orders join orderdetails using(orderNumber) limit 20;
    
````

- We have have some challenges, for example where need to get ```country``` from ```customers.``` That is another join ....


````python
%%sql

select orderNumber, orderDate, customerNumber, productCode, quantityOrdered, priceEach, country
    from
        (select * from orders join orderdetails using(orderNumber)) as a
    join
        customers
    using(customerNumber)
limit 20;
````

- But, we wanted by "region" or "continent." For example, many companies track revenue by region:
    - NA = North America
    - LATAM = Latin America
    - EMEA = Europe, Middle East, Asia
    - APAC = Asia Pacific
    
    
- I downloaded some information that will help.

````python
%%sql

select * from countries limit 10;
````

- We should be able to use this table to map country to continent and then continent to sales region, ... ...

````python
%%sql

SELECT distinct customers.country, countries.* FROM
	customers left join classicmodelsnew.countries
		on country_name=country;
````

- Most of the rows from ```customers``` does not have a match. Let's try something else ... ...

````python
%%sql

SELECT distinct customers.country, countries.* FROM
	customers left join classicmodelsnew.countries
		on country_name like concat("%", country, "%");
````

- Well, sort of, but ... ...

````python
%%sql

SELECT distinct customers.country as customers_country, countries.* FROM
	customers left join classicmodelsnew.countries
		on country_name like concat("%", country, "%")
where
    customers.country in ('Ireland', 'USA', 'Norway', 'Netherlands')
````
