# Impedance Mismatch


- We are to look at ```classicmodels.```


- Let's consider and order form or order printout. We have seen this data tends to be nested.


- But, the relational model is "flat" tables. Let's get all of the data for an order.

````python
%%sql
    select *
        from orders join orderdetails using(orderNumber)
    where customerNumber=363
````

- If I look up the orders for a customer, I expect a list of orders. What I get is a table with
    - One entry for each ```orderDetails```
    - And a lot of duplicate information.
    
    
- I can correct this impedance mismatch if data models with some code.

````python
import pymysql
import json

conn = pymysql.connect(host="localhost",
                       user="dbuser",
                       password="dbuserdbuser",
                       db="classicmodelsnew",
                       cursorclass=pymysql.cursors.DictCursor)


def get_order_details_by_customer(customerNumber=None):
    cur = conn.cursor()

    if customerNumber is not None:
        q = """
            select *
            from orders join orderdetails using(orderNumber)
                where customerNumber=%s order by orderNumber, orderLineNumber
        """
    else:
        q = """
            select *
            from orders join orderdetails using(orderNumber)
                order by orderNumber, orderLineNumber
        """

    res = cur.execute(q, (customerNumber))
    res = cur.fetchall()
    return res


def get_orders_by_customer(customerNumber):
    o_details = get_order_details_by_customer(customerNumber)
    orders = {}

    for o in o_details:
        c_o = orders.get(o["orderNumber"], None)

        if c_o is None:
            c_o = {
                "orderNumber": o["orderNumber"],
                "customerNumber": o["customerNumber"],
                "orderDate": o["orderDate"],
                "requiredDate": o["requiredDate"],
                "shippedDate": o["shippedDate"],
                "status": o["status"],
                "orderDetails": []
            }
        orders[o["orderNumber"]] = c_o

        new_d = {
            "orderLineNumber": o["orderLineNumber"],
            "productCode": o["productCode"],
            "quantityOrdered": o["quantityOrdered"],
            "priceEach": o["priceEach"]
        }
        orders[o["orderNumber"]]["orderDetails"].append(new_d)

    return orders


def load_all_orders():
    orders = get_orders_by_customer(None)
    return orders
````

````python
res = get_order_details_by_customer(363)
````

````python
res
````

- Now make look like a "document order."

````python
orders = get_orders_by_customer(363)
print("Orders = \n", json.dumps(orders, indent=2, default=str))
````

# Do an Update

- Similar in concept to SQL: Takes a selection criteria (WHERE ...) and an update expression.

````python
def get_order_by_number(o_number):
    coll = mongo_conn["classic_models"]["orders"]
    result = coll.find_one(
        {"orderNumber": o_number}
    )
    return result
    
````

````python
o = get_order_by_number(10424)
print("Order = \n", json.dumps(o, indent=2, default=str))
````
